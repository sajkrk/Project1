```python
import os
import numpy as np
import pandas as pd
import csv
import json
import requests
import time
import datetime as dt
```

```python
# data for the year 2016
city="New York City, NY"
year=2016

url=f"https://api.usno.navy.mil/moon/phase?year={year}&loc={city}"
#print(url)

#converting into json format
response=requests.get(url).json()
response
```

```python
# creating a list and a dataframe
date_2016=[]
phase_2016=[]

# using for loop to find the date and phase in each row
for entry in response["phasedata"]:
# date    
    date_2016.append(entry["date"])
#phase    
    phase_2016.append(entry["phase"])
    
moon_2016 = pd.DataFrame({"Date": date_2016, "Phase": phase_2016})
print(moon_2016)
```
